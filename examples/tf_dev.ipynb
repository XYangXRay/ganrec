```python
import numpy as np
import tifffile
import tensorflow as tf
import matplotlib.pyplot as plt
x = tifffile.imread('./test_data/tooth.tiff')
x = np.reshape(x, (1, 181, 400))
x = tf.keras.layers.Conv1D(400, 3, activation='relu', padding = 'same')(x)
x = tf.transpose(x, perm=[0, 2, 1])
x = tf.keras.layers.Conv1D(400, 3, activation='relu', padding = 'same')(x)
x = x.numpy()
```
